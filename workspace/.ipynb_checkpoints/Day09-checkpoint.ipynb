```python
st = input('숫자를 포함한 문자열을 입력하세요 >> ')
a_rst =''
i_rst =''
for i in st:
    if i.isalpha():
        a_rst += i
    elif i.isdigit():
        i_rst += i
print('문자 : ', a_rst)
print('숫자 : ', i_rst)
```

```python
st = input('숫자를 포함한 문자열을 입력하세요 >> ')
a_rst =''
i_rst =''
a_cnt = 0
i_cnt = 0
for i in st:
    if i.isalpha():
        a_rst += i
        a_cnt += 1
    elif i.isdigit():
        i_rst += i
        i_cnt += 1
print(f'문자({a_cnt}) : {a_rst}')
print(f'숫자({i_cnt}) : {i_rst}')

print(f'문자({len(a_rst)}) : {a_rst}')
print(f'숫자({len(i_rst)}) : {i_rst}')
```

```python
st = input('숫자를 포함한 문자열을 입력하세요 >> ')
a_rst =''
i_rst =''
a_cnt = 0
i_cnt = 0
for i in st:
    if i >= 'A' and i <='Z' or i >= 'a' and i <= 'z':
        a_rst += i
        a_cnt += 1
    elif i >= '0' and i <='9':
        i_rst += i
        i_cnt += 1
print(f'문자({a_cnt}) : {a_rst}')
print(f'숫자({i_cnt}) : {i_rst}')
```

```python
# 알고리즘 복습

# 알고리즘은 문제나 과제를 해결하기 위한 처리 절차를 구체적인 순서로 표현한 아이디어나 생각
# 알고리즘 표현 방법은 수도코드(pseudo code), 순서도(flow chart)

# 프로그램 작성 순서
# 기획 - 설계(알고리즘) - 프로그래밍 - 디버그

# 좋은 알고리즘
# 알기 쉽고, 속도가 빠르고, 효율적이고(메모리를 적게사용), 재이용이 쉽다.

# 왜 알고리즘 ??
# 좋은 프로그램 만들수 있고 좋은 프로그램인지 판단할 수 있고
# 프로그램 작성을 효율적으로 할 수 있고 프로그래밍 기술을 향상시키는데 큰 도움이 된다.

# 알고리즘의 조건
# 결과를 얻을 수 있어야 하고 반드시 종료되어야 한다.

# 알고리즘 세가지 기본형
# 1. 순차구조 : 처음부터 순서대로 처리
# 2. 선택구조 : 조건식 if
# 3. 반복구조 : 반복문 for while

# 순서도 flow 차트, 흐름도
# 프로그래밍 언어를 사용하지 않고 알고리즘을 기술하는 방법
# 처리과 절차를 시각적으로 표현한다.
# 자신이 생각하고있는 알고리즘이 올바른지 처리와 절차에 문제가 없는지를
# 눈으로 확인할 수 있고 다른 사람에게도 쉽게 전달할 수 있는 특징이 있다.

# 주의 사항
# 처리 기호나 판단 기호가 많아지면 흐름이 복잡하게 꼬인다.
# 1. 입력은 반드시 위에서 들어가도록 만든다.
# 2. 판단기호로 부터 나오는 흐름선에는 Yes no를 확실하게 기입한다.
# 3. 입력선이나 출력선이 없는 처리 기호는 만들지 않는다.
```
