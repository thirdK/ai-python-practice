```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
html = requests.get('https://movie.naver.com/movie/sdb/rank/rmovie.naver')
soup = bs(html.text)
```

```python
len(soup.select('tbody > tr > td.title'))
movies = soup.select('tbody > tr > td.title')
movies[0]
```

```python
movies[0].select('td > div > a')[0].text
rank = 0
movie_data = []

for movie in movies:
    title = movie.select('td > div > a')[0].text
    rank += 1
    movie_data.append([rank, title])
    
movie_data
df = pd.DataFrame(movie_data, columns = ['순위', '타이틀'])
df
# df.to_excel('movie.xlsx', index = False)
```
