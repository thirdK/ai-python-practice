```python
# 삼각형의 면적을 계산하는 알고리즘

# 삼각형의 면적 = 밑변 * 높이 / 2


# 1.의사코드로 작성

# - 밑변과 높이를 입력
# - area <- base * height / 2
# - area 출력

# 2.플로우차트로 작성
```

```python
# 3.코드로 작성
base = int(input('밑변 >> '))
height = int(input('높이 >> '))

area = base * height / 2
print(area)
```

```python
# 두 데이터의 대소를 판단하는 알고리즘

# # 의사코드
# - 두 데이터를 입력(a,b)
# - if a > b then a출력
# - elif a = b then 같다 출력
# - else then b를 출력

```

```python
#코드로
a = int(input('정수1 >> '))
b = int(input('정수2 >> '))

if a>b:
    print(f'큰 수 : {a}')
elif b>a:
    print(f'큰 수 : {b}')
else:
    print('같다.')
```

```python
#합계값을 계산하는 알고리즘

# 첨자 index  0  1  2  3  4
# 데이터      12 13 11 14 10

```

```python
arr = [12, 13, 11, 14, 10]
total = 0
for i in arr:
    total += i

print(total)
```
