```python
# Linear Search 선형 탐색

# 보관된 데이터를 맨 앞부터 순서대로 원하는 값을 찾을 때까지 탐색한다.
# 탐색 처리는 반복 구조를 사용하여 기술한다.
# 반복 구조에서는 종료 조건을 잊지 말자
```

```python
lst = [1,3,5,7,2,4]

target = int(input('찾을 정수 입력 >> '))
# count = 0

# for i in lst:
#     if target == i:
#         print(f'{count}번째 요소가 일치')
#         break
#     elif count == len(lst)-1:
#         print('찾지 못했습니다.')
#         break
#     count += 1

for i in range(0,len(lst)):
    if lst[i] == target:
        print(f'{i}번째 요소가 일치')
        break
    elif i == len(lst)-1:
        print('찾지 못했습니다.')
        break
```

```python
def linear_search(target, lst):
    for i in range(0, len(lst)):
        if lst[i] == target:
            st = f'{i}번째 요소가 일치'
            return st
    st = '찾지 못했습니다.'
    return st

print(linear_search(6, [4,2,3,6,1])) 
```
